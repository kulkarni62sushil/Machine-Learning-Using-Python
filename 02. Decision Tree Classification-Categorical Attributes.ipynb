## LabelEncoder 

In machine learning, it is difficult to process categorical data. So we have to convert into numeric data using  `LabelEncoder`. `LabelEncoder` encodes labels by assigning them numbers. Thus, if the feature is color with values such as [‘white’, ‘red’, ‘black’, ‘blue’]., using `LabelEncoder` may encode color string label as [0, 1, 2, 3]. 


```python
from sklearn.preprocessing import LabelEncoder
```

```python
# object creation for all categorical data
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
#Create a new numeric variable for these three objects
df['company_n'] = le_company.fit_transform(df['company'])
df['job_n'] = le_job.fit_transform(df['job'])
df['degree_n'] = le_degree.fit_transform(df['degree'])
```

```python
df.head()
```

```python
df1=df.drop(['company','job','degree'],axis='columns')
```

```python
df1.head()
```

```python
#split dataset in feature or training and target variable
from sklearn.model_selection import train_test_split
feature_cols = ['company_n', 'job_n', 'degree_n']
X = df1[feature_cols] # Features
Y = df1.label # Target variable
```

```python
# Split dataset into training set and test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
# Create Decision Tree classifer object
#for decision tree object
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
# Train Decision Tree Classifer
clf = clf.fit(X_train,Y_train)

#Predict the response for test dataset
Y_pred = clf.predict(X_test)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20), dpi=500)
tree.plot_tree(clf,feature_names=feature_cols, class_names=None ,filled=True, rounded=True);
```

```python
# Load libraries
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.rename(columns={"Survived": "label"})
```

```python
df.Age_n = df.Age_n.fillna(df.Age_n.mean())
df.head(6)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Pclass = LabelEncoder()
le_Sex = LabelEncoder()
le_Age = LabelEncoder()
le_Fare = LabelEncoder()
```

```python
df['Pclass_n'] = le_Pclass.fit_transform(df['Pclass'])
df['Sex_n'] = le_Sex.fit_transform(df['Sex'])
df['Age_n'] = le_Age.fit_transform(df['Age'])
df['Fare_n'] = le_Fare.fit_transform(df['Fare'])
```

```python
df.head()
```

```python
df1=df.drop(['Pclass','Sex','Age','Fare'],axis='columns')
df1.head()
```

```python
#split dataset in feature or training and target variable
from sklearn.model_selection import train_test_split
feature_cols = ['Pclass_n', 'Sex_n', 'Age_n','Fare_n']
X = df1[feature_cols] # Features
Y = df1.label # Target variable
```

```python
# Split dataset into training set and test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1) # 80% training and 20% test
```

```python
# Create Decision Tree classifer object
#for decision tree object
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
# Train Decision Tree Classifer
clf = clf.fit(X_train,Y_train)

#Predict the response for test dataset
Y_pred = clf.predict(X_test)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20), dpi=500)
tree.plot_tree(clf,feature_names=feature_cols, class_names=None ,filled=True, rounded=True);
```
